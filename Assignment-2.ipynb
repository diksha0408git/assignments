## Pandas assignment

```python
import pandas as pd
```

```python
data={
    'Name':['Alice','Bob','Charlie','Diana','Eve'],
    'Age':[24,27,22,32,29],
    'Department':['HR','Finance','IT','Marketin','HR'],
    'Salary':[450000,54000,50000,62000,470000]
    }
```

```python
df=pd.DataFrame(data)
```

```python
print('First five row\n',df.head())
```

```python
print('Summary static of age and salary\n',df[['Salary','Age']].describe())
```

```python
print('Average salary of HR department',df[df['Department']=='HR']['Salary'].mean())
```

```python
df['Bonus']=df['Salary']*0.10
df
```

```python
print('Employee btwn age 25 and 30\n',df[(df['Age']>=25) & (df["Age"]<=30)])
```

```python
df.groupby('Department')['Salary'].mean()
```

```python
df.to_csv('new created csv file',index=False)
```

## Matplotlib assignment

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1, 2, 3, 4, 5]
y=[10, 15, 25, 30, 50]
```

```python
plt.figure()
plt.plot(x, y)
plt.title("Line plot")
plt.xlabel('X value')
plt.ylabel('Y value')
plt.grid(True)
plt.show()
```

```python
student=['John','Jam','Aalice','Bbo']
marks=[75,85,60,90]
```

```python
plt.figure()
plt.bar(student, marks)
plt.title("Marks score by students")
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

```python
data=np.random.randint(1,101,1000)
```

```python
plt.figure()
plt.hist(data)
plt.title("Histogram")
plt.xlabel('value')
plt.ylabel('Frequency')
plt.show()
```

```python
regions=['North America','Europe','Asia','Others']
revenue=[45,25,20,10]
```

```python
plt.pie(revenue,labels=regions,autopct='%1.1f%%')
plt.show()
```
