```python
dict={
    101:{'name':'Satya','age':27,'department':'HR','salary':50000}
    }
```

```python
def add(emp_id,name,age,department,salary):
    if emp_id in dict:
        print('employee already exists')
    else:
        dict[emp_id]={'name':name,'age':age,'department':department,'salary':salary}
        print("Employee added successfully")
```

```python
add(102,'Ritu',28,'Tech',60000)
```

```python
def display():
    for emp_id in dict:
        print(f"Employee id:{emp_id}")
        print(f"Name:{dict[emp_id]['name']}")
        print(f"Age:{dict[emp_id]['age']}")
        print(f"Department:{dict[emp_id]['department']}")
        print(f"Salary:{dict[emp_id]['salary']}")
        print("-"*30)
    else:
        print("All employee display successfully")
```

```python
display()
```

```python
def search():
    emp_id = int(input("Enter Employee ID to search: "))

    if emp_id in dict:
        emp = dict[emp_id]
        print("\nEmployee Found:")
        print(f"Name : {emp['name']}")
        print(f"Age: {emp['age']}")
        print(f"Department: {emp['department']}")
        print(f"Salary: {emp['salary']}\n")
    else:
        print("Employee not found.\n")
```

```python
search()
```

```python
def main_menu():
    while True:
        print("Employee Management System")
        print("1. Add Employee")
        print("2. View All Employees")
        print("3. Search for Employee")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == '1':
            add()
        elif choice == '2':
            display()
        elif choice == '3':
            search()
        elif choice == '4':
            print("Thank you for using the Employee Management System!")
            break
        else:
            print("Invalid choice. Please try again.\n")
```

```python
main_menu()
```
